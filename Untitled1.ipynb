```python
# gere x e y
x = digitos.data
y = digitos.target
# formato de x e y
print(x.shape)
print(y.shape)
```

```python
#CLASSIFICAÇÃO KNN EM PYTHON - PRÉ-PROCESSAMENTO E NORMALIZAÇÃO
```

```python
# divisão em dados de treino e de teste
x_treino, testeData, y_treino, testeLabels = train_test_split(x, y, test_size = 0.30, random_state = 101)
# divisão dos dados de treino em dados de treino e dados de validação
treinoData, validacaoData, treinoLabels, validacaoLabels = train_test_split(x_treino, y_treino, test_size = 0.1, random_state = 84)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# imprimindo o numero de exemplos ( observações)em cada dataset
print("Número de observações em treinoData: {}".format(len(treinoData)))
print("Número de observações em treinoLabels: {}".format(len(treinoLabels)))
print("Número de observações em validacaoData: {}".format(len(validacaoData)))
print("Número de observações em validacaoLabels: {}".format(len(validacaoLabels)))
print("Número de observações em testeData: {}".format(len(testeData)))
print("Número de observações em testeLabels: {}".format(len(testeLabels)))
```

```python
# normalização dos dados pela média
# cálculo da média do dataset de treino
x_norm = np.mean(x, axis = 0)
# normalização dos dados de treino e de teste
x_treino_norm = treinoData - x_norm
x_teste_norm = testeData - x_norm
x_validacao_norm = validacaoData - x_norm
```

```python
# shape dos datasets
print(x_treino_norm.shape)
print(x_teste_norm.shape)
print(x_validacao_norm.shape)
```

```python
 # Classificação KNN em python - testando o melhor valor de K
```

```python
# range de valores de k que iremos testar
kvals = range(1, 30, 2)
# lista vazia para receber as acuracias
acuracias = []
# loop em todos os valores de k para testae cada um deles
for k in kvals:
  # treinando o modelo KNN com cada valor em k
  modeloKNN = KNeighborsClassifier(n_neighbors = k)
  modeloKNN.fit(treinoData, treinoLabels)
  # avaliando o  modelo e atualizando a lista de acuracias
  score = modeloKNN.score(validacaoData, validacaoLabels)
  print(" com valor de k = %d, a acuracia é = %.2f%%" %(k, score * 100))
  acuracias.append(score)
  # obtendo o valor de k que apresenta a maior acuracia
  i = np.argmax(acuracias)
  print (" o valor de k = %d alcançou a mais alta acuracia de %.2f%% nos dados de validação!" % (kvals[i], acuracias[i] * 100))
```

```python
# Classificação KNN em python - construção e treinamento do modelo KNN
```

```python
# criando a versão final do modelo com maior valor de k
modeloFinal = KNeighborsClassifier(n_neighbors = kvals[i])
# treinamento do modelo
modeloFinal.fit(treinoData, treinoLabels)


```

```python
# Classificação KNN em python- previsões com dados de teste e avaliação do modelo
```

```python
# previsões com os dados de teste
predictions = modeloFinal.predict(testeData)
# perfomance do modelo nos dados de teste
print("avaliação do  modelo nos dados de teste")
print(classification_report(testeLabels, predictions))
print("confusion matrix")
print(confusion_matrix(testeLabels, predictions))

```
